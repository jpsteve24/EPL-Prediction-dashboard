```python
home_elo_list = []
away_elo_list = []
```

```python
K = 30

for index, row in df.iterrows():

    home_team = row["HomeTeam"]
    away_team = row["AwayTeam"]

    home_elo = elo_ratings[home_team]
    away_elo = elo_ratings[away_team]

    home_elo_list.append(home_elo)
    away_elo_list.append(away_elo)

    expected_home = 1 / (1 + 10 ** ((away_elo - home_elo) / 400))
    expected_away = 1 - expected_home

    if row["FTR"] == "H":
        actual_home = 1
        actual_away = 0

    elif row["FTR"] == "A":
        actual_home = 0
        actual_away = 1

    else:
        actual_home = 0.5
        actual_away = 0.5

    elo_ratings[home_team] = home_elo + K * (actual_home - expected_home)

    elo_ratings[away_team] = away_elo + K * (actual_away - expected_away)
```

```python
home_defense_strength = (
    df.groupby("HomeTeam")["FTAG"].mean()
    / avg_away_goals
)

away_defense_strength = (
    df.groupby("AwayTeam")["FTHG"].mean()
    / avg_home_goals
)
```

```python
def predict_match_probability(

    home_team,

    away_team,

    simulations=10000
):

    expected_home_goals = (

        home_attack_strength[home_team]

        * away_defense_strength[away_team]

        * avg_home_goals
    )

    expected_away_goals = (

        away_attack_strength[away_team]

        * home_defense_strength[home_team]

        * avg_away_goals
    )

    home_wins = 0
    draws = 0
    away_wins = 0

    for i in range(simulations):

        simulated_home_goals = np.random.poisson(
            expected_home_goals
        )

        simulated_away_goals = np.random.poisson(
            expected_away_goals
        )

        if simulated_home_goals > simulated_away_goals:

            home_wins += 1

        elif simulated_home_goals < simulated_away_goals:

            away_wins += 1

        else:

            draws += 1

    home_probability = home_wins / simulations

    draw_probability = draws / simulations

    away_probability = away_wins / simulations

    print("Home Win Probability:",
          round(home_probability, 3))

    print("Draw Probability:",
          round(draw_probability, 3))

    print("Away Win Probability:",
          round(away_probability, 3))
```
